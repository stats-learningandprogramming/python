## <a href = "https://plotnine.readthedocs.io/en/stable/#">About plotnine</a>

<b> A Grammar of Graphics for Python </b>
plotnine is an implementation of a <i>grammar of graphics</i> in Python, it is based on ggplot2. The grammar allows user to compose plots by explicitly mapping data to the visual objects that make up the plot.
<br>
Plotting with a grammar is powerful, it makes custom (add otherwise complex) plots easy to think about and then create, while the simple plots remain simple.
<br>

### History and objective

The plotnine started as an effort to improve the scaling functionality in ggpy formally known as "ggplot for python". It was part of a larger goal to evolve the package into one that supported a full plotting grammer. It turned out that to ahve a <i>grammer of graphics</i> system we had to just about start a new. <br>
<br>
The complete guide to what constitutes a "Gammar of Graphics" is Leland Wilkinson's book <i>The Grammar of Graphics</i>. To create ggplot2 Hadley Wickham came up with an interpretation termed <i>A layered grammar of graphics</i>. Core to the interpretation is a crucial plot building pipeline in ggplot2 that we adopted for plotnine. <br>
<br>
The R programming language has a rich statistical ecosystem that ggplot2 taps into with ease. In plotnine we have done our best to integrate with the rest of the scientific ecosystem. Though we feel we could do more on that integration, notwithstanding language differences, users familiar with ggplot2 should be comfortable and productive with plotnine.

### Built with

- matplotlib - Plotting backend.
- pandas - Data handling.
- mizani - Scales framework.
- statsmodels - For various statistical computations.
- scipy - For various statistical computation procedures.

### Example

```python
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap
from plotnine.data import mtcars

(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm')
 + facet_wrap('~gear'))
```
